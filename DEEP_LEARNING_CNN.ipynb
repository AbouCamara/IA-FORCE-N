#Historique

Mais en 2012, une révolution se produit : lors de la compétition annuelle de vision par ordinateur ILSVRC, un nouvel algorithme de Deep Learning explose les records ! Il s'agit d'un réseau de neurones convolutif appelé AlexNet.

**AlexNet** a été développé par des chercheurs de Toronto en 2012, mais est largement inspiré du réseau de neurones convolutif LeNet, développé en 1998 par le chercheur français Yann LeCun. LeNet est donc le pionnier, et AlexNet celui qui a créé l'enthousiasme autour du Deep Learning !<br>
Aujourd'hui, les réseaux de neurones convolutifs, aussi appelés CNN ou ConvNet pour Convolutional Neural Network, sont toujours les modèles les plus performants pour la classification d'images.

**Définition**
Les réseaux de neurones convolutionnels (CNN, Convolutional Neural Networks en anglais) sont une classe de réseaux de neurones utilisée principalement pour la reconnaissance d'images et d'autres tâches de traitement de données visuelles. Ils sont appelés "convolutionnels" car ils utilisent une opération mathématique appelée convolution pour filtrer les données d'entrée.

Les CNN sont construits à partir de couches de neurones qui sont connectées de manière à apprendre des caractéristiques spécifiques dans les données d'entrée. La première couche de neurones d'un CNN prend en entrée l'image elle-même et applique une série de filtres convolutionnels pour extraire des caractéristiques de cette image. Ces filtres sont généralement de petites matrices de nombres appelées "noyaux" ou "kernels" qui sont glissées sur l'image et multipliées par les valeurs des pixels de l'image pour produire une nouvelle image appelée "carte de caractéristiques".

Les couches suivantes de neurones sont généralement des couches de "max pooling", qui réduisent la dimension de l'image en sélectionnant les valeurs maximales dans chaque sous-région de l'image. Cela permet de capturer les caractéristiques les plus importantes de l'image tout en réduisant la quantité de données à traiter.

Les couches ultérieures du réseau de neurones convolutionnel sont généralement des couches "pleinement connectées", qui sont similaires aux couches de neurones traditionnelles utilisées dans les réseaux de neurones "fully connected" (FC). Ces couches utilisent les caractéristiques extraites par les couches précédentes pour effectuer des prédictions sur les données d'entrée.

Les réseaux de neurones convolutionnels sont largement utilisés dans de nombreuses tâches de traitement d'images, telles que la reconnaissance de caractères manuscrits, la reconnaissance de visages, la détection d'objets et la segmentation d'images. Ils sont également utilisés dans de nombreuses autres tâches de traitement de données, comme la reconnaissance de la parole et la traduction automatique.

**Mode Fonctionnement** <br>
**Couche de convolution**
Supposons que nous avons une image de 10 x 10 pixels et que nous voulons appliquer un filtre convolutionnel de 3 x 3 pixels sur cette image. Le filtre est initialement rempli de valeurs aléatoires, qui seront ajustées au fil de l'entraînement du réseau de neurones.

Pour appliquer le filtre sur l'image, nous commençons en haut à gauche de l'image et déplaçons le filtre sur l'image en "glissant" horizontalement et verticalement jusqu'à ce que nous ayons couvert toute l'image. À chaque étape, nous multiplions chaque valeur du filtre par la valeur du pixel correspondant de l'image et nous additionnons tous ces produits pour obtenir un nouveau nombre appelé "score". Ce score est enregistré dans une nouvelle image appelée "carte de caractéristiques".

Max pooling <br>
Le max pooling est une couche utilisée dans les réseaux de neurones convolutionnels pour réduire la dimension des données tout en conservant les caractéristiques les plus importantes. Il s'agit d'une opération de downsampling qui divise l'image en sous-régions et sélectionne la valeur maximale de chaque sous-région.<br>
Le max pooling est utilisé dans les réseaux de neurones convolutionnels pour réduire la dimension des données tout en conservant les caractéristiques les plus importantes de l'image. Cela permet de réduire la quantité de données à traiter dans les couches ultérieures du réseau, ce qui peut accélérer le temps de calcul et améliorer la performance du réseau.

**Remarque**<br>
L'opération donne deux types de résultats : l'un dans lequel la dimension de la caractéristique convoluée est réduite par rapport à l'entrée, et l'autre dans lequel la dimension est augmentée ou reste la même. Ceci est fait en appliquant **Valid Padding ** dans le cas du premier, ou **Same Padding ** dans le cas du second.<br>

Lorsque nous augmentons l'image 5x5x1 en une image 6x6x1 et que nous appliquons ensuite le noyau 3x3x1 sur celle-ci, nous constatons que la matrice convoluée s'avère être de dimensions 5x5x1. D'où le nom - ** Same Padding.**

D'autre part, si nous effectuons la même opération sans remplissage, nous obtenons une matrice dont les dimensions sont celles du noyau (3x3x1) lui-même - **Valid Padding**.

```python
from keras.datasets import fashion_mnist
```

```python
(X_train,y_train),(X_test,y_test)=fashion_mnist.load_data()

X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
X_train[1]
```

```python
import matplotlib.pyplot as plt

plt.imshow(X_train[1])
plt.show()
```

```python
from tensorflow.keras.utils import to_categorical

y_train_enc=to_categorical(y_train,10)
y_test_enc=to_categorical(y_test,10)
X_train_norm = X_train /255

X_test_norm = X_test / 255
```

```python
X_train_norm= X_train_norm.reshape(X_train.shape[0],28,28,1)
X_test_norm=X_test_norm.reshape(X_test.shape[0],28, 28, 1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D, Flatten,Input,MaxPool2D,Dropout


def get_model(input_shape, classes):
  model=Sequential()

  model.add(Input(input_shape)) #couche d'entrée

  model.add(Conv2D(filters=16, kernel_size=(3,3), activation="relu")) #couche conv
  model.add(MaxPool2D((2,2)))
  model.add(Conv2D(filters=8, kernel_size=(3,3), activation="relu"))
  model.add(MaxPool2D((2,2)))

  model.add(Flatten()) #transforme la matrice en vecteur

  model.add(Dense(units=32, activation="relu"))
  model.add(Dropout(0.5))
  model.add(Dense(units=24, activation="relu"))
  model.add(Dense(units=20, activation="relu"))
  model.add(Dropout(0.3))
  model.add(Dense(units=16, activation="relu"))
  model.add(Dense(units=classes, activation="softmax")) #couche de sortie

  return model
```

```python
model=get_model((28,28,1), 10)

model.summary()
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model)
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy",metrics=['accuracy'])
```

```python
history=model.fit(X_train_norm,y_train_enc,epochs=50,batch_size=64, validation_data=(X_test_norm,y_test_enc))
```

```python
import matplotlib.pyplot as plt
hist=history.history

plt.figure(figsize=(12,8))
plt.plot(hist['val_loss'],c="r")
plt.plot(hist['loss'])
plt.show()
```

```python
import matplotlib.pyplot as plt
hist=history.history

plt.figure(figsize=(12,8))
plt.plot(hist['val_accuracy'],c="r")
plt.plot(hist['accuracy'])
plt.show()
```

#Le dropout
Le dropout est une technique de régularisation utilisée dans les réseaux de neurones profonds (deep learning). Elle consiste à mettre au hasard certains neurones "hors service" pendant l'entraînement, en les mettant à zéro. Cela signifie que ces neurones ne contribuent pas à la prédiction finale et ne reçoivent pas de gradient durant l'entraînement.

Le but du dropout est de réduire l'overfitting, c'est-à-dire l'aptitude du modèle à trop bien s'adapter aux données d'entraînement, au détriment de sa capacité à généraliser sur de nouvelles données. En mettant au hasard certains neurones hors service, le modèle est obligé de s'adapter à des configurations différentes à chaque itération d'entraînement, ce qui le rend moins dépendant des caractéristiques individuelles des données d'entraînement.

Le taux de dropout est généralement défini par l'utilisateur et peut être ajusté en fonction de la performance du modèle. Un taux de dropout élevé peut entraîner une réduction de l'overfitting, mais peut également entraîner une diminution de la performance du modèle sur les données d'entraînement.

exemple: model.add(Dropout(0.5))

Cela signifie que pendant l'entraînement, la moitié des neurones de ces couches seront mis au hasard à zéro.

Vous pouvez également utiliser le dropout avec d'autres types de couches, comme les couches Conv2D dans les réseaux de neurones convolutionnels. Dans ce cas, le dropout s'applique à chaque canal de sortie de la couche de convolution.

Il est important de noter que le dropout ne s'applique qu'à l'entraînement du modèle et non à la phase de test. Pour cette raison, il est recommandé de désactiver le dropout lors de l'inférence, en utilisant la méthode model.evaluate() ou model.predict().

#BATCH NORMALISATION
La normalisation par lots (batch normalisation) est une technique utilisée pour améliorer la performance et la stabilité des réseaux de neurones profonds (deep learning). Elle consiste à normaliser les activations des couches du modèle en utilisant les moyennes et écarts-types des données de chaque batch d'entraînement.
<br>
La normalisation par lots a plusieurs avantages :<br>

Elle peut réduire l'overfitting en rendant le modèle moins dépendant des caractéristiques individuelles des données d'entraînement.
Elle peut accélérer la convergence du modèle en rendant les données plus faciles à traiter pour les couches suivantes.
Elle peut améliorer la stabilité du modèle en réduisant l'explosion et l'évanouissement des gradients.
Il est important de noter que la normalisation par lots ne s'applique qu'à l'entraînement du modèle et non à la phase de test. Pour cette raison, il est recommandé de désactiver la normalisation par lots lors de l'inférence, en utilisant la méthode model.evaluate() ou model.predict().
